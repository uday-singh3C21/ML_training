```python
import numpy as np 
import pandas as pd 
from sklearn import linear_model
```

```python
advertising = pd.read_csv('advertising.csv')
advertising.head(4)
```

```python
reg = linear_model.LinearRegression()
reg.fit(advertising[['TV','Radio','Newspaper']],advertising.Sales)
```

```python
reg.predict([[230.1,37.8,69.2]])
```
